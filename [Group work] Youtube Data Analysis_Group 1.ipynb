# 문제1 - 년도별 평균 조회수

```python
table1['year'] = table1.publish_time.str.split('-').str[0]
table1
```

```python
year_view_count=table1.groupby('year').views.count()
year_view_count
```

```python
year_view_sum=table1.groupby('year').views.sum()
year_view_sum
```

```python
table_result=year_view_sum/year_view_count
table_result
```

```python
import matplotlib.pyplot as plt
```

```python
table_result.plot(title="year_view_mean")
```

#  문제2 - 카테고리별 평균 조회수

```python
avg_views= table1.groupby('category_id').views.mean() 
avg_views
```

```python
avg_views_result=avg_views.sort_values()
avg_views_result
#27:education, 25:뉴스,politcs, 26:스타일, 15:pets&animals, 19:travel
```

```python
avg_views_result=avg_views_result.to_frame()
avg_views_result.columns=['avg_views']
```

```python
avg_views_result.plot.bar()
#10번 뮤직, 29번: nonprofit&activism #1번:영화  #24 entertainment #17 sports
```

# 문제 5 - 나라별 카테고리 별 개수와 나라별 카테고리 별 조회수를 구하기

```python
cat1=US.groupby('category_id').category_id.count() 
cat2=GB.groupby('category_id').category_id.count() 
```

```python
cat1=cat1.sort_values(ascending=False)
cat1
```

```python
cat2=cat2.sort_values(ascending=False)
cat2
```

```python
avg_views1= US.groupby('category_id').views.mean() 
avg_views2= GB.groupby('category_id').views.mean()
```

```python
avg_views1
avg_views1=avg_views1.sort_values(ascending=False)
```

```python
avg_views1
```

```python
avg_views2
avg_views2=avg_views2.sort_values(ascending=False)
```

```python
avg_views2
```

```python
result1 = pd.concat([cat1, avg_views1], axis=1)
result1
```

```python
result1.columns=['category_id_sum','category_id_views_sum']
result1
```

```python
result1.sort_values(by="category_id_sum", ascending=False)
```

```python
result1.corr()
```

```python
result2 = pd.concat([cat2, avg_views2], axis=1)
result2.columns=['category_id_sum','category_id_views_sum']
result2.sort_values(by="category_id_sum", ascending=False)
```

```python
result2.corr()
```
